# Amazon Reviews Scraping

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
base_url = 
next_pages = "/pageNumber={pgnb}"

# Nam


url = "https://www.amazon.in/AVITA-Celeron-Storage-NS12T5IN021P-Charcoal/product-reviews/B09HKBR46P"
```

# Individual Chunks

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text)
soup
```

```python
alfa = soup.find_all(name='div',attrs={'class':'a-section review aok-relative'})
print(len(alfa))
alfa[0]
```

```python
gama = []

for beta in alfa:
    one_record = {}
    # Reviewer Name
    one_record['Reviewer Name'] = beta.find_all('div',{'class':'a-profile-content'})[0].text
    # Rating
    one_record['Rating'] = beta.find_all('i',{'data-hook':'review-star-rating'})[0].text
    # Title
    one_record['Review Title'] = beta.find_all('a',{'data-hook':'review-title'})[0].text.strip()
    # Review
    one_record['Review Text'] = beta.find_all('span',{'data-hook':'review-body'})[0].text.strip()
    gama.append(one_record)
    
df = pd.DataFrame(gama)
print(df.shape)
df.head()
```

# Get all Reviews

```python
pg_no = 1
gama = []

while True:
    complete_url = url + next_pages.format(pgnb=pg_no)
    response = requests.get(url)
    status = response.status_code
    
    if status == 200:
        soup = BeautifulSoup(response.text)
        alfa = soup.find_all(name='div',attrs={'class':'a-section review aok-relative'})
        
        for beta in alfa:
            one_record = {}
            # Reviewer Name
            one_record['Reviewer Name'] = beta.find_all('div',{'class':'a-profile-content'})[0].text
            # Rating
            one_record['Rating'] = beta.find_all('i',{'data-hook':'review-star-rating'})[0].text
            # Title
            one_record['Review Title'] = beta.find_all('a',{'data-hook':'review-title'})[0].text.strip()
            # Review
            one_record['Review Text'] = beta.find_all('span',{'data-hook':'review-body'})[0].text.strip()
            gama.append(one_record)        
    else:
        break
    pg_no += 1
    
df = pd.DataFrame(gama)
print(df.shape)
df.head()
```
